# Кейс "Подготовка данных для анализа"

## Вариант 3

Дан датасет, который содержит информацию о потреблении алкоголя по континентам.

1. Загрузите датасет в блокнот из файла drink.csv

2. Выведите информацию о датасете:

    * количество строк и колонок,
    * информацию о типах данных, нулевых и ненулевых значениях в колонках (признаках).

3. Создайте в ячейках таблицы которые отвечают на следующие вопросы:

    * На каком континенте в среднем пьют больше пива?
    * Для каждого континента выведите статистику потребления вина.
    * Выведите среднее потребления алкоголя на каждом континенте для каждого столбца.
    * Выведите среднее, минимальное и максимальное значения для потребления спиртных напитков.

4. Решение представьте в виде ячейки с кодом на языке Python. Решение может содержать несколько строчек. Используемый пакет — `pandas`.
