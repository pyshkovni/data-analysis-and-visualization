# Кейс "Подготовка данных для анализа"

## Вариант 1

Даны статистические данные по демографии в США за 2015 год.

1. Загрузите датасет в блокнот из файла acs2015_census_tract_data.csv.gz

2. Выведите информацию о датасете:

    * количество строк и колонок;
    * информацию о типах данных, нулевых и ненулевых значениях в колонках (признаках).

3. Создайте сводную таблицу, для которой выполняются следующие требования:

    * таблица содержит штаты, где больше половины населения является темнокожим и процент безработицы больше 50 процентов;
    * в ячейках таблица содержит количество мужчин и женщин в штатах согласно условиям выше.

1. Решение представьте в виде ячейки с кодом на языке Python. Решение может содержать несколько строчек. Используемый пакет — `pandas`.
