# Кейс "Подготовка данных для анализа"

## Вариант 5

Даны исходные данные, которые содержат информацию о сборных чемпионата европы по футболу за 2012 год.

1. Загрузите два датасета в блокнот из файла Euro_2012_stats_TEAM.xlsx для листов 'result' и 'stat'

2. Выведите информацию о датасете:

    * количество строк и колонок,
    * информацию о типах данных, нулевых и ненулевых значениях в колонках (признаках).

3. Создайте в ячейках таблицы которые отвечают на следующие вопросы:

    <!-- * Склейте две таблицы по стране сборной -->
    * Выведите список стран, сборные которых забили более 5 голов включительно и название которых начинается с буквы 'G'
    * Выведите таблицу, которая включает название страны из полученного списка выше, точность попадания и точность передач 

4. Решение представьте в виде ячейки с кодом на языке Python. Решение может содержать несколько строчек. Используемый пакет — `pandas`.
